```python
import sklearn

print(sklearn.__version__)
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
# 모듈 정리 완료
```

```python
# 데이터 준비
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
print(dir(digits))
```

```python
digits_data = digits.data

print(digits_data.shape)  # shape는 배열의 형상정보를 출력
```

```python
# 예시 데이터 확인
digits_data[0]
```

```python
# feature Data 지정하기
digits_feature_names = digits.feature_names
```

```python
digits.target_names
```

```python
# digits에 어떤 정보가 담겨 있는지 확인
digits.keys()
```

```python
# target 정보 확인
digits_label = digits.target
print(digits_label.shape)
digits_label
```

```python
# 라벨이름은 target_names 를 확인 가능
digits.target_names
```

```python
# 나머지 변수 확인

print(digits.DESCR)
```

```python
 digits.feature_names
```

# 데이터 학습 시키기 

```python
import pandas as pd

print(pd.__version__)
```

```python
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
digits_df

# digits 데이터 자료형 변환 및 문제지 출제
```

```python
# 정답 설정 하기

digits_df["label"] = digits.target
digits_df
```

```python
# sklearn.model _selection 패키지의 train_test_split을 활용

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                    digits_label,
                                                    test_size=0.2,
                                                    random_state=7)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

```python
# X_train 부터 X_test까지 만들어진 데이터셋 확인

X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape

# 설정한 대로 20%만큼의 데이터 테스트 데이터셋에 나머지는 트레인 데이터셋에 들어간거 확인
```

```python
# y 확인하기

y_train, y_test
```

```python
# DecisionTreeClassifier 사용하기

from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
print(decision_tree._estimator_type)
```

```python
# 한줄로 정리하기

decision_tree.fit(X_train, y_train)
```

# 머신러닝 모델 비교하기

```python
# decision_tree로 테스트 데이터로 예측하기

y_pred = decision_tree.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Random Forest로 모델 평가하기

from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=21)

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn import svm
svm_model = svm.SVC()

print(svm_model._estimator_type)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)
```
